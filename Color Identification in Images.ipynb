# image color detector which identifies all the colors in an image or video.


```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image=cv2.imread('color.jpg')
```

```python
#Converting BGR image of colours to RGB
RGB=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.figure(figsize=(20,8))
plt.imshow(RGB)
```

```python
#Converting image of colours from RGB to HSV
HSV=cv2.cvtColor(RGB,cv2.COLOR_RGB2HSV)
```

```python
LHSV=np.array([25,150,50])
UHSV=np.array([35,255,255])
```

```python
#Detecting Yellow colour form the image 
mask=cv2.inRange(HSV,LHSV,UHSV)
plt.figure(figsize=(20,8))
plt.imshow(mask)
```

```python
res=cv2.bitwise_and(RGB,RGB,mask=mask)
plt.figure(figsize=(20,8))
plt.imshow(res)
```

```python
image2=cv2.imread('sun.jpg')
```

```python
plt.figure(figsize=(20,8))
plt.imshow(image2)
```

```python
#Converting Image into RGB format
RGB=cv2.cvtColor(image2,cv2.COLOR_BGR2RGB)
plt.figure(figsize=(20,8))
plt.imshow(RGB)
```

```python
#Converting RGB image into HSV
HSV=cv2.cvtColor(RGB,cv2.COLOR_RGB2HSV)
```

```python
#Identifying the GREEN color in the image
LGREEN = np.array([25,150,50])  #Lower HSV values of GREEEN color
UGREEN = np.array([85,255,255]) #Upper HSV values of GREEN color
```

```python
mask=cv2.inRange(HSV,LGREEN,UGREEN)
plt.figure(figsize=(20,8))
plt.imshow(mask)
```

```python
#Removing mask and showing only GREEN color
temp=cv2.bitwise_and(RGB,RGB,mask=mask)
plt.figure(figsize=(20,8))
plt.imshow(temp)
```
